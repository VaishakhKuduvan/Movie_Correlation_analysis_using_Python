```python
# Import libraries
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline 
matplotlib.rcParams['figure.figsize'] = (12,6)
```

```python
 # reading the dataset
df = pd.read_csv(r'X:\Data Analyst Portfolio\Python\movies.csv')
df.head()
```

```python
# We need to see if we have any missing data
# Let's loop through the data and see if there is anything missing

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Data Types for our columns

print(df.dtypes)
```

```python
df["budget"] = df["budget"].fillna(0).astype('int64')
df["gross"] = df["gross"].fillna(0).astype('int64')
```

```python
df.head()
```

```python
# Creating the correct year column
df['year_correct'] = df['released'].str.extract(r'(\d{4})')
df.head()
```

```python
df.drop_duplicates()
```

```python
# Order our Data a little bit to see

df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

df['budget_m'] = df['budget'] / 1_000_000
df['gross_m'] = df['gross'] / 1_000_000

plt.scatter(df['budget_m'], df['gross_m'], alpha=0.5)
plt.xlabel('Budget (Millions)')
plt.ylabel('Gross (Millions)')
plt.title('Budget vs Gross')

# Tick every 500 million — adjust to taste
plt.gca().yaxis.set_major_locator(ticker.MultipleLocator(500))
plt.show()
```

```python
df.head()
```

### Plotting the scatterplot

```python
# Plot the budget vs gross using seaborn
sns.regplot(x = "budget", y = 'gross', data = df, scatter_kws= {'color' : "red"}, line_kws={'color' : 'blue'})

```

### Looking at the correlations now (using 3 different methods)

```python
# Looking at the correlations now

# Pearson,  Kendall, Spearman 
df.corr(numeric_only=True, method='pearson')
```

```python
df.corr(numeric_only=True, method='kendall')
```

```python
df.corr(numeric_only=True, method='spearman')
```

#### Insights : 
High correlation between budget and gross. Pearson and Spearman agree (~0.75/0.72), but Kendall is lower (0.54). 
This suggests the relationship is mostly monotonic but not perfectly linear, and there are likely outliers 
(big-budget flops or surprise hits) pulling Pearson up slightly.

```python
correlation_matrix = df.corr(numeric_only=True,method='spearman')
sns.heatmap(correlation_matrix, annot=True)
plt.xlabel('Movie features')
plt.ylabel('Movie features')
plt.title('Correlation  Matrix for numeric Features')
plt.show()
```
